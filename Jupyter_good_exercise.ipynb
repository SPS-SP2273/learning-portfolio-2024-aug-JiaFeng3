```python
%%timeit 
squares = [i**2 for i in range(10)]
```

```python
print(squares)
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

```python
print(squares)
```

The first number is the mean amount of time taken to execute the code.(in microseconds)
The second number is the standard deviation of time taken to execute the code.(in nano seconds)
The code was ran 100,000 times for 7 runs to achieve the readings or values.
